```python
import pandas as pd

df = pd.read_csv("employee_salary_dataset.csv.csv")

df.head()
df.info()
df.describe()
```

```python
df.groupby("Department")["Monthly_Salary"].mean()
```

```python
df["Department"].value_counts()
```

```python
df[["Experience_Years","Monthly_Salary"]].corr()
```

```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
sns.barplot(x="Department", y="Monthly_Salary",data=df)
plt.title("Average Monthly Salary by Department")
plt.xlabel("Department")
plt.ylabel("Monthly Salary")

plt.show()
```

### Insight: Salary by Department

From the bar chart, we can observe the following:

• The Marketing department has the highest average monthly salary among all departments.

• The Finance department has the lowest average monthly salary.

• Operations and HR departments have moderate salary ranges compared to other departments.

• This analysis helps understand how salary distribution varies across departments in the organization.
